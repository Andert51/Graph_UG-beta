## 1. Introduction & Features

**GraphUG** is an open-source, interactive desktop application for mathematical computing, symbolic algebra, and data visualization. It provides a MATLAB/Octave-like experience with a modern, themeable interface.

### Key Capabilities

| Category | Highlights |
|---|---|
| **Numeric Computing** | 200+ built-in functions (NumPy-powered), GPU acceleration via CuPy |
| **Symbolic Algebra** | Derivatives, integrals, limits, series, factoring, solving (SymPy) |
| **2-D Plotting** | Line, scatter, bar, histogram, polar, parametric, vector fields, heatmaps |
| **3-D Visualization** | Surfaces, wireframes, scatter3d, parametric surfaces, bar3d (PyOpenGL) |
| **Linear Algebra** | Determinants, inverses, eigenvalues, SVD, QR, LU, Cholesky, RREF |
| **Signal Processing** | FFT, IFFT, convolution, windowing functions |
| **Customizable UI** | 5 built-in themes, dockable panels, toolbar, settings dialog |

### Technology Stack

- **GUI:** PySide6 (Qt 6) with QDockWidget panels
- **Parsing:** Lark grammar → custom AST evaluator
- **Numerics:** NumPy (with optional CuPy GPU backend)
- **Symbolics:** SymPy
- **2-D Graphs:** PyQtGraph
- **3-D Graphs:** PyQtGraph OpenGL (PyOpenGL)

### Requirements

- Python **3.12+**

```bash
# 1. Clone the repository
git clone https://github.com/Andert51/Graph_UG-beta.git
cd Graph_UG-beta

# 2. Create a virtual environment
python -m venv venv

# 3. Activate it
# Windows:
venv\Scripts\activate
# macOS / Linux:
# source venv/bin/activate

# 4. Install dependencies
pip install -r requirements.txt
```

## 4. Basic Syntax

GraphUG uses a custom expression language parsed by a Lark grammar.

### Operators (Precedence: lowest → highest)

| Precedence | Operator | Example |
|:---:|---|---|
| 1 | Ternary `? :` | `x > 0 ? x : -x` |
| 2 | Pipe `\|>` | `[3,1,2] \|> sort` |
| 3 | Logical `or` | `a or b` |
| 4 | Logical `and` | `a and b` |
| 5 | Logical `not` | `not a` |
| 6 | Comparison | `==  !=  <  >  <=  >=` |
| 7 | Add / Sub | `a + b`, `a - b` |
| 8 | Mul / Div / Mod | `a * b`, `a / b`, `a % b` |
| 9 | Power (right-assoc.) | `2^3^2` = `2^(3^2)` = 512 |
| 10 | Unary | `-x`, `+x` |
| 11 | Postfix | `v[i]` (index), `f(x)` (call) |

### Comments

Lines starting with `#` are ignored:

```
# This is a comment
x = 5  # inline comments are NOT supported
```

### Multi-Statement Lines

Separate statements with `;`:

```
a = 1; b = 2; a + b
```

### String Literals

Used for symbolic expressions and labels:

```
diff("x^3", "x")
xlabel("Time (s)")
```

## 6. Vectors & Matrices

### Vectors

Vectors are 1-D NumPy arrays created with square brackets:

```
v = [1, 2, 3]           # → array([1., 2., 3.])
w = [4.5, -2.1, 0]      # → array([4.5, -2.1, 0.])
```

**Indexing** (0-based):
```
v[0]    # → 1.0
v[2]    # → 3.0
```

**Operations:**
```
v + w           # element-wise addition
v * 2           # scalar multiplication
dot(v, w)       # dot product → 32.0
cross(v, w)     # cross product (3-D vectors)
norm(v)         # magnitude → 3.7416...
```

### Matrices

Matrices are 2-D NumPy arrays. Use `;` to separate rows:

```
m = [1, 2; 3, 4]               # 2×2 matrix
A = [1, 2, 3; 4, 5, 6; 7, 8, 9]  # 3×3 matrix
```

**Operations:**
```
det(m)              # determinant → -2.0
inv(m)              # inverse
transpose(m)        # transpose
matmul(m, m)        # matrix multiplication
eigvals(m)          # eigenvalues
trace(m)            # sum of diagonal → 5.0
rank(m)             # matrix rank → 2
```

### Constructors

```
eye(3)              # 3×3 identity matrix
zeros(2, 3)         # 2×3 matrix of zeros
ones(4)             # vector of four 1s
diag([1, 2, 3])     # diagonal matrix
rand(3, 3)          # random uniform matrix
linspace(0, 10, 5)  # [0, 2.5, 5, 7.5, 10]
```

## 8. Plotting & Visualization

### Basic 2-D Plots

#### Line Plot
```
x = linspace(0, 2*pi, 200)
plot(x, sin(x))
```

#### Scatter Plot
```
scatter([1, 2, 3, 4], [1, 4, 9, 16])
```

#### Bar Chart
```
bar([5, 10, 15, 8, 12])
```

#### Histogram
```
data = randn(1000)
hist(data, 30)
```

#### Vector Arrow
```
vector(3, 4)                    # from origin
vector(1, 1, 3, 4)              # from (1,1)
```

### Overlay Mode (Hold)

By default, each plot clears the canvas. Use `hold()` to overlay:

```
hold(1)
x = linspace(0, 2*pi, 200)
plot(x, sin(x))
plot(x, cos(x))
plot(x, tan(x))
hold(0)
```

### Function Plotting (Symbolic)

Plot a math expression directly from a **string**:

```
fplot("sin(x)")                         # default range [-10, 10]
fplot("x^2 - 3*x + 1", -5, 5)           # custom range
fplot("log(x)", 0.01, 10, 500)           # custom range + 500 points
```

### Polar Plot

```
polar("1 + sin(t)")                     # cardioid
polar("cos(3*t)")                        # 3-petal rose
polar("t", 0, 6*pi)                      # spiral
```

### Parametric Curves

```
# 2-D parametric
parametric("cos(t)", "sin(t)")           # circle
parametric("cos(3*t)", "sin(2*t)", 0, 2*pi)  # Lissajous

# 3-D parametric
parametric3d("cos(t)", "sin(t)", "t/10", 0, 20)  # helix
```

### Derivative & Integral Visualization

```
plotderiv("x^3 - 2*x")                  # function + derivative overlay
plotintegral("x^2", 0, 2)               # function + shaded integral area
tangentline("x^2", 1)                   # function + tangent line at x=1
```

### 3-D Surfaces

```
surface("sin(x)*cos(y)")                # default [-5, 5] × [-5, 5]
surface("x^2 + y^2", -3, 3, -3, 3)      # custom domain
wireframe("sin(sqrt(x^2 + y^2))")
```

### Heatmaps & Contours

```
heatmap("exp(-(x^2 + y^2))")            # 2-D color map
contour("x^2 + y^2", -3, 3, -3, 3, 8)   # contour lines
implicit("x^2 + y^2 - 1")               # implicit curve (circle)
```

### Vector Fields

```
slopefield("y - x")                     # ODE direction field
vectorfield("-y", "x")                  # rotation vector field
```

### Special 2-D Plots

```
stem(x, sin(x))                          # stem (lollipop) plot
step(x, floor(x))                        # staircase plot
area(x, sin(x))                          # filled area plot
errorbar([1,2,3], [4,5,6], [0.5,0.3,0.4])  # error bars
pie([30, 20, 50])                        # pie chart
```

### Log-Scale Plots

```
x = logspace(0, 3, 50)
logplot(x, x^2)                          # log-log
semilogx(x, linspace(0, 10, 50))         # log x-axis
semilogy(linspace(0, 5, 50), exp(linspace(0, 5, 50)))  # log y-axis
```

### 3-D Specialty Plots

```
scatter3d(rand(100), rand(100), rand(100))
surfparam("cos(u)*sin(v)", "sin(u)*sin(v)", "cos(v)")  # sphere
bar3d([1,2,3], [1,2,3], [5,10,8])       # 3-D bars
```

### Canvas Labels & Grid

## 9. Symbolic Algebra & Calculus

All symbolic functions accept **string expressions** and return symbolic results via SymPy.

### Algebra

```
simplify("sin(x)^2 + cos(x)^2")        # → 1
factor("x^2 - 5*x + 6")                 # → (x - 2)*(x - 3)
expand("(x + 1)^3")                      # → x³ + 3x² + 3x + 1
```

### Derivatives

```
diff("x^3", "x")                         # → 3*x²
diff("sin(x)", "x", 2)                   # 2nd derivative → -sin(x)
partial("x^2*y", "x")                    # partial derivative → 2*x*y
```

### Integrals

```
integrate("x^2", "x")                    # indefinite → x³/3
defint("x^2", "x", 0, 1)                 # definite → 1/3
```

### Limits & Series

```
limit("sin(x)/x", "x", 0)               # → 1
series("exp(x)", "x", 0, 6)              # Taylor series
taylor("sin(x)", "x", 0, 7)              # Taylor polynomial
```

### Summation & Product

```
summation("k^2", "k", 1, 10)             # Σ k² from 1 to 10
product("k", "k", 1, 5)                   # Π k from 1 to 5 = 120
```

### Equation Solving

```
solve("x^2 - 4", "x")                    # → [-2, 2]
nsolve("cos(x) - x", "x", 1)             # numerical → 0.7390...
```

### Vector Calculus (Symbolic)

```
gradient("x^2 + y^2 + z^2", "x,y,z")     # → [2x, 2y, 2z]
divergence("x^2, y^2, z^2", "x,y,z")     # → 2x + 2y + 2z
curl("-y, x, 0", "x,y,z")                # → [0, 0, 2]
laplacian("x^2 + y^2", "x,y")             # → 4
```

### Integral Transforms

```
laplace("exp(-a*t)", "t", "s")            # → 1/(a + s)
invlaplace("1/(s + a)", "s", "t")          # → exp(-a*t)
```

### Symbolic Linear Algebra

```
rref("[[1,2,3];[4,5,6]]")                # row-reduced echelon form
nullspace("[[1,2];[2,4]]")               # kernel of matrix
colspace("[[1,2];[3,4]]")                # column space
```

## 11. Statistics & Signal Processing

### Statistics

```
data = randn(1000)

mean(data)                    # arithmetic mean
median(data)                  # median
std(data)                     # standard deviation
var(data)                     # variance
min(data)                     # minimum
max(data)                     # maximum
percentile(data, 95)          # 95th percentile
corrcoef(x, y)                # correlation matrix
cov(x, y)                     # covariance matrix
```

### Reductions & Cumulative

```
sum(v)       prod(v)       cumsum(v)     cumprod(v)
argmin(v)    argmax(v)     cummin(v)     cummax(v)
movmean(v, 5)              # moving average (window=5)
```

### Signal Processing / FFT

```
# Generate a signal
t = linspace(0, 1, 256)
signal = sin(2*pi*50*t) + 0.5*sin(2*pi*120*t)

# Frequency analysis
fft(signal)                   # Fast Fourier Transform
rfft(signal)                  # Real FFT (magnitude)
fftfreq(256, 1/256)           # frequency bins
fftshift(fft(signal))         # center zero-frequency
ifft(fft(signal))             # inverse FFT (reconstruction)

# Convolution
convolve([1, 2, 3], [0, 1, 0.5])  # 1-D convolution

# Window functions
hamming(64)     hanning(64)     blackman(64)     kaiser(64, 14)
```

### Polynomials

```
polyfit(x, y, 2)              # polynomial fit (degree 2)
polyval([1, 0, -1], 2)        # evaluate x² - 1 at x=2 → 3
roots([1, 0, -1])             # roots of x² - 1 → [-1, 1]
linreg(x, y)                  # linear regression
trapz(y, x)                   # numerical integration
```

## 12. Advanced Features

### Pipe Operator

Pass a result to a function using `|>`:

```
[3, 1, 4, 1, 5] |> sort       # → [1, 1, 3, 4, 5]
[1, 2, 3, 4, 5] |> mean       # → 3.0
```

### Ternary Operator

```
x = -5
x > 0 ? x : -x                # → 5 (absolute value)
```

### Bitwise Operations

```
bitand(12, 10)                 # 12 & 10 = 8
bitor(12, 10)                  # 12 | 10 = 14
bitxor(12, 10)                 # 12 ^ 10 = 6
bitnot(12)                     # ~12
shl(1, 4)                      # 1 << 4 = 16
shr(16, 2)                     # 16 >> 2 = 4
```

### Complex Numbers

```
z = complex(3, 4)              # 3 + 4i
real(z)                        # → 3
imag(z)                        # → 4
abs_complex(z)                 # → 5
phase(z)                       # → 0.9272... (radians)
conj(z)                        # → 3 - 4i
```

### Set Operations

```
union([1,2,3], [3,4,5])        # → [1, 2, 3, 4, 5]
intersect([1,2,3], [2,3,4])    # → [2, 3]
setdiff([1,2,3], [2,3,4])      # → [1]
in1d([1,2,3,4], [2,4])         # → [0, 1, 0, 1]
```

## 15. Examples & Recipes

### Recipe 1: Basic Computation

```
# Simple arithmetic
2^10                           # → 1024
factorial(10)                  # → 3628800
sqrt(2)                        # → 1.4142...

# Variables
r = 5
area = pi * r^2                # → 78.5398...
```

### Recipe 2: Plotting Multiple Functions

```
hold(1)
x = linspace(-2*pi, 2*pi, 300)
plot(x, sin(x))
plot(x, cos(x))
plot(x, sin(x) + cos(x))
xlabel("x")
ylabel("f(x)")
title("Trigonometric Functions")
grid(1)
hold(0)
```

### Recipe 3: Matrix System of Equations

Solve the system:
$$2x + y = 5$$
$$x + 3y = 7$$

```
A = [2, 1; 1, 3]
b = [5, 7]
solve_linear(A, b)
# → [1.6, 1.8]
```

### Recipe 4: Fourier Analysis

```
# Create a signal with two frequencies
t = linspace(0, 1, 256)
signal = sin(2*pi*10*t) + 0.5*sin(2*pi*25*t)

# View in time domain
plot(t, signal)
xlabel("Time (s)")
title("Time Domain")

# View in frequency domain
freqs = fftfreq(256, 1/256)
spectrum = rfft(signal)
plot(freqs, spectrum)
xlabel("Frequency (Hz)")
title("Frequency Spectrum")
```

### Recipe 5: Calculus Visualization

```
# Plot a function and its derivative together
plotderiv("x^3 - 3*x", -3, 3)

# Visualize a definite integral
plotintegral("sin(x)", 0, pi)

# Show tangent line at a point
tangentline("x^2", 2, -1, 5)
```

### Recipe 6: 3-D Surface Exploration

```
# Saddle surface
surface("x^2 - y^2", -3, 3, -3, 3)

# Gaussian bell
surface("exp(-(x^2 + y^2))", -3, 3, -3, 3)

# Parametric sphere
surfparam("cos(u)*sin(v)", "sin(u)*sin(v)", "cos(v)")
```

### Recipe 7: Vector Field Analysis

```
# Rotation field
vectorfield("-y", "x", -3, 3, -3, 3)

# Slope field for dy/dx = y - x
slopefield("y - x", -5, 5, -5, 5)

# Verify: curl of rotation field
curl("-y, x, 0", "x,y,z")   # → [0, 0, 2]
```

### Recipe 8: Statistical Analysis

```
# Generate random data
data = randn(1000)

# Summary statistics
mean(data)
std(data)
median(data)
percentile(data, 25)
percentile(data, 75)

# Visualize distribution
hist(data, 30)
title("Normal Distribution (n=1000)")
```
